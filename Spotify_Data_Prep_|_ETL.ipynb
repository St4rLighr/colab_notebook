```python
spotify_df = pd.read_csv('Most Streamed Spotify Songs 2024.csv',encoding = 'ISO-8859-1', encoding_errors= 'ignore')
spotify_df
```

```python
spotify_df.shape
```

```python
spotify_df.info()
```

```python
#picking columns to drop
spotify_df = spotify_df.drop(['TIDAL Popularity','Soundcloud Streams'],axis=1) #dropping soundcloud cause has more han 50% null
```

```python
spotify_df.info()
```

```python
#adding '_' to all column names
spotify_df.columns = [col.replace(" ", "_") for col in spotify_df.columns]

#convert all column names to lowercase
spotify_df.columns = spotify_df.columns.str.lower()
```

```python
spotify_df.columns
```

```python
spotify_df.info()
```

```python
spotify_df
```

```python
#resetting required columns to date and numeric
spotify_df['release_date'] = pd.to_datetime(spotify_df['release_date'])

spotify_df = spotify_df.map(lambda x: x.replace(',','') if isinstance(x,str) else x)

cols_to_format = list(spotify_df.columns[5:6]) + list(spotify_df.columns[7:])

for col in cols_to_format:
    spotify_df[col] = pd.to_numeric(spotify_df[col], errors='coerce')
    spotify_df[col] = spotify_df[col].fillna(spotify_df[col].mean().round()).astype(int) #filling null numerical columns w/mean
```

```python
#filling null numerical columns w/mean
# num_col = spotify_df.select_dtypes('number')
# spotify_df[num_col.columns] = num_col.fillna(num_col.mean())
```

```python
spotify_df.isnull().sum()
```

```python
#Replacing Values
spotify_df['explicit_track'] = spotify_df['explicit_track'].replace(1, True).replace(0, False)
spotify_df = spotify_df.fillna(value={'artist': 'Unknown'})
```

```python
#removing duplicates
spotify_df.drop_duplicates(keep="first", inplace = True)
```
